```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.3, model_complexity=1)

# Define indices for shoulder, elbow, and wrist landmarks
KEYPOINTS = {
    "LEFT_SHOULDER": 11, "RIGHT_SHOULDER": 12,
    "LEFT_ELBOW": 13, "RIGHT_ELBOW": 14,
    "LEFT_WRIST": 15, "RIGHT_WRIST": 16,
    "LEFT_HIP": 23, "RIGHT_HIP": 24,
    "LEFT_KNEE": 25, "RIGHT_KNEE": 26,
    "LEFT_ANKLE": 27, "RIGHT_ANKLE": 28,
    "LEFT_FOOT_INDEX": 31, "RIGHT_FOOT_INDEX": 32
}

# Function to extract and visualize specific keypoints
def extract_and_draw_selected_keypoints(frame):
    # Process the frame
    results = pose.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    if results.pose_landmarks:
        # Convert landmarks to a list
        landmarks = results.pose_landmarks.landmark

        # Draw selected keypoints and connections
        for key, idx in KEYPOINTS.items():
            landmark = landmarks[idx]
            print(landmark)
            x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])

            # Draw keypoint
            cv2.circle(frame, (x, y), radius=5, color=(0, 255, 0), thickness=-1)

        # Draw connections (shoulder -> elbow -> wrist)
        connections = [
            (KEYPOINTS["LEFT_SHOULDER"], KEYPOINTS["LEFT_ELBOW"]),
            (KEYPOINTS["LEFT_ELBOW"], KEYPOINTS["LEFT_WRIST"]),
            (KEYPOINTS["RIGHT_SHOULDER"], KEYPOINTS["RIGHT_ELBOW"]),
            (KEYPOINTS["RIGHT_ELBOW"], KEYPOINTS["RIGHT_WRIST"]),
        ]

        for start_idx, end_idx in connections:
            start = landmarks[start_idx]
            end = landmarks[end_idx]
            start_point = (int(start.x * frame.shape[1]), int(start.y * frame.shape[0]))
            end_point = (int(end.x * frame.shape[1]), int(end.y * frame.shape[0]))

            # Draw connection
            cv2.line(frame, start_point, end_point, color=(255, 0, 0), thickness=2)
    else:
        print("Failed")

    return frame

# Load the frame
frame = cv2.imread(r"E:\PosePerfect\Frames\biceps_curl\augmented_augmented_biceps_curl_video_23_frame10.jpg")

# Extract and draw the specific keypoints
output_frame = extract_and_draw_selected_keypoints(frame)

# Display the frame
cv2.imshow("Selected Keypoints Visualization", output_frame)

# Wait until a key is pressed, then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()

# Release MediaPipe Pose
pose.close()
```
