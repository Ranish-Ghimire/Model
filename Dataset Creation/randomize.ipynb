```python
import numpy as np

num_samples = 1612
seq_length = 15
height, width, channels = 224, 224, 3

X_data = np.memmap('X_combined_1612.dat', dtype='float32', mode='r+', shape=(num_samples, seq_length, height, width, channels))
y_data = np.memmap('y_combined_1612.dat', dtype='int32', mode='r+', shape=(num_samples,))
print(X_data.shape)
```

```python
indices = np.arange(num_samples)
np.random.shuffle(indices)
```

```python
X_final = np.memmap('X_final_1612.dat', dtype='float32', mode='w+', shape=(num_samples, seq_length, height, width, channels))
y_final = np.memmap('y_final_1612.dat', dtype='int32', mode='w+', shape=(num_samples,))

X_final[:] = X_data[indices]
y_final[:] = y_data[indices]
```

```python
X_final.flush()
y_final.flush()
```
