```python
import numpy as np

num_samples = 274
seq_length = 15
height, width, channels = 224, 224, 3

X_data = np.memmap('X_combined_remain.dat', dtype='float32', mode='r+', shape=(num_samples, seq_length, height, width, channels))
y_data = np.memmap('y_combined_remain.dat', dtype='int32', mode='r+', shape=(num_samples,))
print(X_data.shape)
```

```python
indices = np.arange(num_samples)
np.random.shuffle(indices)
```

```python
X_final = np.memmap('X_final_remain.dat', dtype='float32', mode='w+', shape=(num_samples, seq_length, height, width, channels))
y_final = np.memmap('y_final_remain.dat', dtype='int32', mode='w+', shape=(num_samples,))

for i, idx in enumerate(indices):
    X_final[i] = X_data[idx]         
    y_final[i] = y_data[idx]
```

```python
X_final.flush()
y_final.flush()
```

```python
X_data = np.memmap('X_final_15k.dat', dtype='float32', mode='r+', shape=(num_samples, seq_length, height, width, channels))
y_data = np.memmap('X_final_15k.dat', dtype='int32', mode='r+', shape=(num_samples,))

print(f"{X_data.shape}")
```

```python
import matplotlib.pyplot as plt
import cv2

# Convert the frame from BGR to RGB
frame_rgb = X_data[1][14][:, :, ::-1]  # Reverses the last channel (BGR -> RGB)

# Display the corrected frame
plt.imshow(frame_rgb)
plt.axis('off')
plt.show()
```
