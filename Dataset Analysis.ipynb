```python
from matplotlib import pyplot as plt
import os
import cv2
import numpy as np
```

```python
biceps_curl_count = 0
biceps_curl_path = r"E:/barbell_biceps_curl_videos"
for _ in os.listdir(biceps_curl_path):
    biceps_curl_count += 1

biceps_curl_count
```

```python
squat_count = 0
squat_path = r"E:/squat_videos"
for _ in os.listdir(squat_path):
    squat_count += 1

squat_count
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5, 5))

x_axis = ['biceps_curl', 'squat']
y_axis = [biceps_curl_count, squat_count]

bars = plt.bar(x_axis, y_axis, color='maroon', width=0.4)

for bar in bars:
    yval = bar.get_height() 
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, 
             str(int(yval)), ha='center', va='bottom', fontsize=10)

plt.xlabel("Class")
plt.ylabel("Number of videos")
plt.title("Total Videos")

plt.show()
```

```python
def get_video_duration(video_path):
    video = cv2.VideoCapture(video_path)
    fps = video.get(cv2.CAP_PROP_FPS)
    frame_count = video.get(cv2.CAP_PROP_FRAME_COUNT)
    duration = frame_count / fps
    video.release()
    return duration

paths = [r"E:/barbell_biceps_curl_videos", r"E:/squat_videos"]
total_durations = []
for path in paths:
    total_duration = 0
    for video in os.listdir(path):
        video_path = os.path.join(path, video)
        total_duration += get_video_duration(video_path)
    total_durations.append(total_duration)
    
total_durations
```

```python
total_frame_biceps = total_durations[0] * 30
total_frame_squats = total_durations[1] * 30

print(f'Total frames in biceps curl: {int(total_frame_biceps)}')
print(f'Total frames in squats: {int(total_frame_squats)}')
```

```python
import matplotlib.pyplot as plt

number_of_frames = [total_frame_biceps, total_frame_squats]
exercise_type = ['biceps_curl', 'squats']

bars = plt.bar(exercise_type, number_of_frames, color='maroon', width=0.4)

for bar in bars:
    yval = bar.get_height() 
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, 
             str(int(yval)), ha='center', va='bottom', fontsize=10)

plt.xlabel("Class")
plt.ylabel("Number of Frames")
plt.title("Total Frames before clean-up")

plt.show()
```
